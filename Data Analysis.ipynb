```
# library import
from pyspark.sql import SparkSession
```

```
# Tables Load as Dataframe
df1 = spark.read.table("BASE_EMPILHADAV2") 
```

```
# Libraries import 
import pandas as pd
import matplotlib.pyplot as plt
!pip install seaborn
import seaborn as sns
```

```
# Database filter for 2019 year 
df_filtrado2019 = df1.filter((df1.NU_ANO_CENSO == 2019))
soma_alunos = df_filtrado2019.groupBy("TP_MODALIDADE_ENSINO").sum("QT_ING")
soma_alunos.show()
```

```

# Group by TP_MODALIDADE_ENSINO and calculate QT_ING sum with Spark
soma_alunos_spark = df_filtrado2019.groupBy('TP_MODALIDADE_ENSINO').sum('QT_ING')

# Transform to pandas dataframe
df_agrupado = soma_alunos_spark.toPandas()

# Bars graphy criation
plt.figure(figsize=(10, 6))
plt.bar(df_agrupado['TP_MODALIDADE_ENSINO'], df_agrupado['sum(QT_ING)'], color='blue')

# Title and lable addition
plt.title('An√°lise 2019')
plt.xlabel('Tipo de Modalidade de Ensino')
plt.ylabel('Total de QT_ING')
plt.grid(False)

# Formatting for thousands
from matplotlib.ticker import FuncFormatter
formatter = FuncFormatter(lambda x, _: '{:,.0f}'.format(x/1000)) 
plt.gca().yaxis.set_major_formatter(formatter)
plt.tight_layout()

# Label replacement to improve data understandment. If 'TP_MODALIDADE_ENSINO' = 1, then it is "Presencial", if 'TP_MODALIDADE_ENSINO' = 2, then it is "EAD". 
plt.xticks(ticks=[1, 2], labels=['Presencial', 'EaD'])

plt.show()
```
